```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

# Configuraci n t cnica para entorno Docker
options = Options()
options.add_argument("--headless") # Ejecucion sin interfaz grafica
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Inicializar el navegador
driver = webdriver.Chrome(options=options)

try: 
    # 1. Acceso al sitio
    driver.get("https://tradingeconomics.com/")
    time.sleep(3) # Pausa obligatoria para carga de JavaScript
    
    # 2. Captura de datos (ejemplo)
    items = driver.find_elements(By.CLASS_NAME, "#matrix")
    for i in items:
        print(i.text)

finally:
    # 3. Cierre de s e s i n (Crucial para no colapsar la RAM de Docker)
    driver.quit()
```
