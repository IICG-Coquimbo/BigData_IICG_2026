```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

def iniciar_navegador():
    options = Options()
    options.binary_location = "/usr/bin/google-chrome"
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")

    return webdriver.Chrome(
        service=Service(ChromeDriverManager().install()),
        options=options
    )

URL_OBJETIVO = "https://www.investing.com/currencies/"

SELECTOR_CONTENEDOR = "table tbody tr"
SELECTOR_DATO_A = "td:nth-child(2)"
SELECTOR_DATO_B = "td:nth-child(3)"

driver = iniciar_navegador()
dataset_final = []

try:
    print("Conectando...")
    driver.get(URL_OBJETIVO)
    time.sleep(5)

    elementos = driver.find_elements(By.CSS_SELECTOR, SELECTOR_CONTENEDOR)
    print(f"Se encontraron {len(elementos)} registros")

    for item in elementos:
        try:
            nombre = item.find_element(By.CSS_SELECTOR, SELECTOR_DATO_A).text
            valor = item.find_element(By.CSS_SELECTOR, SELECTOR_DATO_B).text

            valor_limpio = "".join(filter(str.isdigit, valor))

            dataset_final.append({
                "divisa": nombre,
                "precio": valor_limpio,
                "status": 1.0
            })
        except:
            continue

    if dataset_final:
        df = pd.DataFrame(dataset_final)
        df.to_csv("divisas_investing.csv", index=False)
        print(df.head())
    else:
        print("No se capturaron datos")

finally:
    driver.quit()
    print("Navegador cerrado")
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

def iniciar_navegador():
    options = Options()
    options.binary_location = "/usr/bin/google-chrome"
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")

    return webdriver.Chrome(
        service=Service(ChromeDriverManager().install()),
        options=options
    )

URL_OBJETIVO = "https://www.x-rates.com/table/?from=USD&amount=1"

SELECTOR_CONTENEDOR = "table.tablesorter.ratesTable tbody tr"
SELECTOR_DATO_A = "td:nth-child(1)"
SELECTOR_DATO_B = "td:nth-child(2)"

driver = iniciar_navegador()
dataset_final = []

try:
    print("Conectando...")
    driver.get(URL_OBJETIVO)
    time.sleep(3)

    elementos = driver.find_elements(By.CSS_SELECTOR, SELECTOR_CONTENEDOR)
    print(f"Se encontraron {len(elementos)} registros")

    for item in elementos:
        try:
            moneda = item.find_element(By.CSS_SELECTOR, SELECTOR_DATO_A).text
            valor = item.find_element(By.CSS_SELECTOR, SELECTOR_DATO_B).text

            dataset_final.append({
                "moneda": moneda,
                "valor": valor,
                "status": 1.0
            })
        except:
            continue

    if dataset_final:
        df = pd.DataFrame(dataset_final)
        df.to_csv("divisas_xrates.csv", index=False)
        print(df.head())
    else:
        print("No se capturaron datos")

finally:
    driver.quit()
    print("Navegador cerrado")
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

# CONFIGURACIÓN
options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(options=options)

datos_totales = []

# "Paginación" → distintas monedas base
monedas = ["USD", "EUR", "CLP", "GBP"]

try:
    for moneda in monedas:
        print(f"Procesando base: {moneda}")

        url = f"https://www.x-rates.com/table/?from={moneda}&amount=1"
        driver.get(url)
        time.sleep(3)

        filas = driver.find_elements(By.CSS_SELECTOR, "table.tablesorter tbody tr")

        for fila in filas:
            try:
                divisa = fila.find_element(By.CSS_SELECTOR, "td:nth-child(1)").text
                tasa = fila.find_element(By.CSS_SELECTOR, "td:nth-child(2)").text

                datos_totales.append({
                    "base": moneda,
                    "divisa": divisa,
                    "tipo_cambio": tasa
                })

            except:
                continue

    df = pd.DataFrame(datos_totales)
    df.to_csv("divisas_xrates.csv", index=False)

    print(df.head())
    print(f"\nTotal registros: {len(df)}")

finally:
    driver.quit()
    print("Navegador cerrado")
```
