```python
import requests
from bs4 import BeautifulSoup

# Configuración de entrada
url = "https://www.investing.com/currencies/eur-usd"
headers = {"User-Agent": "Mozilla/5.0"}

# 1. Petición al servidor
respuesta = requests.get(url, headers=headers)

# 2. Convertir el HTML en un objeto de Python
sopa = BeautifulSoup(respuesta.text, 'html.parser')

# 3. Buscar los datos (según lo que inspeccionaste)
resultados = sopa.find_all('td', class_='pid-1-last')

# 4. Mostrar en pantalla
for dato in resultados:
    print(f"Precio encontrado: {dato.text.strip()}")
```
