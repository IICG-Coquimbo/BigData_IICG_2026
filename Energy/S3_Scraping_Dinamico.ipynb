```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# 1. Configuración básica del navegador
options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(options=options)

# 2. Variables para los datos
datos_totales = []
paginas_a_recorrer = 3 

try:
    # Entramos a la sección de documentos o noticias del Coordinador
    driver.get("https://www.coordinador.cl/category/documentos/")
    
    for i in range(paginas_a_recorrer):
        print(f"Leyendo página {i + 1}...")

        # Esperamos a que carguen los títulos de los documentos
        # En esta página, los títulos suelen estar en etiquetas <h3> o <article>
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.TAG_NAME, "h3"))
        )

        # Capturamos los títulos que hay en la vista actual
        elementos = driver.find_elements(By.TAG_NAME, "h3")
        for el in elementos:
            texto = el.text.strip()
            if texto:
                datos_totales.append(texto)

        # Lógica para cambiar de página
        try:
            # XPATH común para el botón "Siguiente" o la flecha de paginación
            xpath_next = "//a[contains(@class, 'next')] | //a[contains(text(), 'Siguiente')]"
            
            boton_siguiente = WebDriverWait(driver, 5).until(
                EC.element_to_be_clickable((By.XPATH, xpath_next))
            )
            
            # Hacemos clic y esperamos un poco a que cargue el contenido nuevo
            driver.execute_script("arguments[0].click();", boton_siguiente)
            time.sleep(2) 

        except:
            print("No se encontró más paginación.")
            break

    print(f"\nFinalizado. Se recolectaron {len(datos_totales)} títulos.")

except Exception as e:
    print(f"Error: {e}")

finally:
    # Cerramos el navegador para no dejar procesos colgados
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(options=options)

datos_totales = []
paginas_a_recorrer = 3 

try:
    driver.get("https://www.coordinador.cl/category/documentos/")
    
    for i in range(paginas_a_recorrer):
        print(f"Leyendo página {i + 1}...")

        # Esperamos a los artículos (usamos una clase más específica del sitio)
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CLASS_NAME, "entry-title"))
        )

        elementos = driver.find_elements(By.CLASS_NAME, "entry-title")
        for el in elementos:
            datos_totales.append(el.text.strip())

        # Intentamos el cambio de página con el selector exacto del sitio
        try:
            # Este es el selector específico para el botón "Siguiente" del Coordinador
            xpath_next = "//a[contains(@class, 'next')]"
            
            boton_siguiente = WebDriverWait(driver, 5).until(
                EC.element_to_be_clickable((By.XPATH, xpath_next))
            )
            
            driver.execute_script("arguments[0].click();", boton_siguiente)
            time.sleep(4) # Le damos más tiempo de carga entre páginas

        except:
            print("No se encontró más paginación o fin de lista.")
            break

    print(f"\n✅ Finalizado. Se recolectaron {len(datos_totales)} títulos.")

except Exception as e:
    print(f"❌ Error: {e}")

finally:
    driver.quit()
```
