```python
from pymongo import MongoClient
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By # Necesario para buscar
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# 1. Conectar al contenedor de Mongo
client = MongoClient('mongodb://mongodb:27017/')

# 2. Crear (o usar) la base de datos y la colección
db = client['proyecto_bigdata']
coleccion = db['datos_scraping']

print("Conexión exitosa a MongoDB")

# --- PASO INTERMEDIO: DEFINIR 'productos' ---
options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.binary_location = "/usr/bin/google-chrome"
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

driver.get("https://www.coordinador.cl/category/documentos/") # O la URL de tu proyecto
time.sleep(5)

# AQUÍ DEFINIMOS 'productos' PARA QUE NO DE ERROR
productos = driver.find_elements(By.TAG_NAME, "h3") 


# Dentro del bucle de productos:
for p in productos:
    # Creamos un diccionario (documento JSON)
    documento = {
        "texto_crudo": p.text,
        "fecha_captura": time.strftime("%H:%M:%S - %d/%m/%Y"), # <-- CAMBIAMOS EL ORDEN AQUÍ
        "grupo": "G1_Ecommerce"
    }

    # Insertar en MongoDB
    coleccion.insert_one(documento)
    print(f"Insertado con hora: {documento['fecha_captura']}")

driver.quit()
print("✅ Proceso finalizado.")
```
