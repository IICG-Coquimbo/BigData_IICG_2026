```python
import requests
from bs4 import BeautifulSoup

# Fuente alternativa estática para el grupo de Energía
url = "https://www.revistaei.cl/categoria/electromov/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
}

print(f"Conectando a la fuente de Energía...")
respuesta = requests.get(url, headers=headers)

if respuesta.status_code == 200:
    sopa = BeautifulSoup(respuesta.text, 'html.parser')
    
    # Buscamos los títulos de las noticias de energía
    noticias = sopa.find_all('h2', class_='entry-title')
    
    if not noticias:
        print("No se encontraron noticias con esos criterios.")
    else:
        print(f"--- Últimas Noticias de Mercados Energéticos ---")
        for noticia in noticias:
            # Extraemos el texto y el link
            titulo = noticia.text.strip()
            link = noticia.find('a')['href']
            print(f"Título: {titulo}")
            print(f"Link: {link}")
            print("-" * 30)
else:
    print(f"Error al conectar: {respuesta.status_code}")
```
