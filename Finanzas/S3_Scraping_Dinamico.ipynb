```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

options = Options()
options.binary_location = "/usr/bin/google-chrome"
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

driver.get("https://www.investing.com/currencies/")
time.sleep(5)

print("Página cargada correctamente")
```

```python
from selenium.webdriver.common.by import By

datos = []

elementos = driver.find_elements(By.TAG_NAME, "a")

print(f"Se encontraron {len(elementos)} elementos")

for item in elementos[:20]:
    texto = item.text.strip()
    
    if texto:
        datos.append({
            "nombre": texto,
            "categoria": "Finanzas"
        })

print(datos[:5])
```

```python
import pandas as pd

df = pd.DataFrame(datos)
df.to_csv("finanzas_dinamico.csv", index=False)

print(df.head())
```

```python
driver.quit()
print("Navegador cerrado")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

driver.get("https://www.investing.com/currencies/")
time.sleep(5)

datos = []

for i in range(3):
    print(f"Procesando página {i+1}")
    
    elementos = driver.find_elements(By.TAG_NAME, "a")
    
    for item in elementos:
        texto = item.text.strip()
        if texto:
            datos.append(texto)
    
    try:
        boton = driver.find_element(By.XPATH, "//a[contains(text(),'Next')]")
        boton.click()
        time.sleep(3)
    except:
        print("No hay más páginas")
        break

driver.quit()

print(f"Total datos: {len(datos)}")
```
