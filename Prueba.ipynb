```python
# PRUEBA DE SALUD DEL ENTORNO
import platform
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

print(f"1. Sistema: {platform.system()} | Arquitectura: {platform.machine()}")

options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.binary_location = "/usr/bin/google-chrome"

try:
    driver = webdriver.Chrome(
        service=Service(ChromeDriverManager().install()),
        options=options
    )
    print("2. Driver y Chrome sincronizados.")
    driver.get("https://www.google.com")
    print(f"3. Conexión a Internet: {driver.title}")
    driver.quit()
    print("4. Entorno listo para trabajar.")
except Exception as e:
    print(f"ERROR EN EL RESET: {e}")
    
```

```python
from pymongo import MongoClient

print("Paso 1: conectar a Mongo")

client = MongoClient(
    "mongodb://mongodb:27017/",
    serverSelectionTimeoutMS=3000  # 👈 evita que se quede pegado
)

print("Paso 2: crear base y colección")
db = client["bigdata"]
coleccion = db["reset_semana4"]

print("Paso 3: insertar documento de prueba")

try:
    resultado = coleccion.insert_one({
        "estado": "ok",
        "mensaje": "Mongo funcionando en Semana 4"
    })

    print("Paso 4: documento insertado")
    print("ID:", resultado.inserted_id)

except Exception as e:
    print("❌ ERROR:", e)
```
