```python
import requests
from bs4 import BeautifulSoup

url = "http://books.toscrape.com"
headers = {"User-Agent": "Mozilla/5.0"}

respuesta = requests.get(url, headers=headers)
sopa = BeautifulSoup(respuesta.text, 'html.parser')

resultados = sopa.find_all('h3')

todos_los_datos = []

for dato in resultados:
    titulo = dato.find('a')['title']
    
    info = {
        "titulo": titulo,
        "grupo": "Books"
    }
    
    todos_los_datos.append(info)

print("Cantidad de datos:", len(todos_los_datos))
print(todos_los_datos[:3])
```

```python
precios = sopa.find_all('p', class_='price_color')

for i in range(len(resultados)):
    titulo = resultados[i].find('a')['title']
    precio = precios[i].text
    
    info = {
        "titulo": titulo,
        "precio": precio
    }
    
    print(info)
```
