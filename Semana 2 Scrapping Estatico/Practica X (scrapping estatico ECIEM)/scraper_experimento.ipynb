```python
import requests
from bs4 import BeautifulSoup

url = "https://www.eciem.cl/index.php/anuncios/pregrados/2/Ingeniera-en-Informacin-y-Control-de-Gestin"
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

titulo = soup.title.string
print(f"He capturado el título: {titulo}")

avisos = soup.find_all('div', class_='item-publicaciones')
for aviso in avisos:
    fecha = aviso.find('span', class_='fechas')
    titulo = aviso.find('h3')
    print(f"Fecha: {fecha.get_text(strip=True)} | Título: {titulo.get_text(strip=True)}")
```

```python
ultimo = avisos[0]

fecha   = ultimo.find('span', class_='fechas').get_text(strip=True)
titulo  = ultimo.find('h3').get_text(strip=True)
enlace  = ultimo.find('a')['href'] if ultimo.find('a') else 'Sin enlace'
cuerpo  = ultimo.find('p').get_text(strip=True) if ultimo.find('p') else 'Sin texto'

print(f"Fecha:  {fecha}")
print(f"Título: {titulo}")
print(f"Enlace: {enlace}")
print(f"Texto:  {cuerpo}")
```

```python
año_buscado = "2026"

for aviso in avisos:
    fecha = aviso.find('span', class_='fechas').get_text(strip=True)
    if año_buscado in fecha:
        titulo = aviso.find('h3').get_text(strip=True)
        enlace = aviso.find('a')['href'] if aviso.find('a') else 'Sin enlace'
        cuerpo = aviso.find('p').get_text(strip=True) if aviso.find('p') else 'Sin texto'
        
        print(f"Fecha:  {fecha}")
        print(f"Título: {titulo}")
        print(f"Enlace: {enlace}")
        print(f"Texto:  {cuerpo}")
        print("---")
```

```python
import json
from datetime import datetime

resultados = []

año_buscado = "2026"

for aviso in avisos:
    fecha = aviso.find('span', class_='fechas').get_text(strip=True)
    if año_buscado in fecha:
        titulo = aviso.find('h3').get_text(strip=True)
        enlace = aviso.find('a')['href'] if aviso.find('a') else 'Sin enlace'
        cuerpo = aviso.find('p').get_text(strip=True) if aviso.find('p') else 'Sin texto'
        
        resultados.append({
            "fecha": fecha,
            "titulo": titulo,
            "enlace": enlace,
            "cuerpo": cuerpo
        })

# Guardar en archivo JSON
with open('avisos_eciem.json', 'w', encoding='utf-8') as f:
    json.dump(resultados, f, ensure_ascii=False, indent=4)

print(f"Se guardaron {len(resultados)} avisos en avisos_eciem.json")
```
