```python
1 import requests
2 from bs4 import BeautifulSoup
3
4 # Configuraci n de entrada
5 url = "PON_AQUI_LA_URL_DE_TU_CASO"
6 headers = {"User-Agent": "Mozilla/5.0"} # Para que no nos bloqueen
7
8 # 1. Petici n al servidor
9 respuesta = requests.get(url, headers=headers)
10
11 # 2. Convertir el HTML en un objeto de Python
12 sopa = BeautifulSoup(respuesta.text, ’html.parser’)
13
14 # 3. Buscar los datos (Ejemplo: buscar todos los t tulos)
15 # Cambia ’h2’ y ’nombre-de-la-clase’ por lo que inspeccionaste
16 resultados = sopa.find_all(’h2’, class_=’nombre-de-la-clase’)
17
18 # 4. Mostrar en pantalla
19 for dato in resultados:
20 print(f"Encontrado: {dato.text.strip()}")
```
