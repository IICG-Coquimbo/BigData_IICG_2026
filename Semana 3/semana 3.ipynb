```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

# ==========================================================
# 1. MOTOR DE NAVEGACIÓN
# ==========================================================
def iniciar_navegador():
    options = Options()
    options.binary_location = "/usr/bin/google-chrome"
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--disable-gpu")
    options.add_argument("user-agent=Mozilla/5.0")

    return webdriver.Chrome(
        service=Service(ChromeDriverManager().install()),
        options=options
    )

# ==========================================================
# 2. CONFIGURACIÓN (TU CASO REAL)
# ==========================================================

URL_OBJETIVO = "https://seguimientoejecucionobras.coordinador.cl/"

# Lo que tú encontraste en el inspector
SELECTOR_CONTENEDOR = "tr.tr-custom"
SELECTOR_DATO_A = "td.text-left"
SELECTOR_DATO_B = "td.text-center"

# ==========================================================
# 3. SCRAPING
# ==========================================================
driver = iniciar_navegador()
dataset_final = []

try:
    print("Conectando a la página...")
    driver.get(URL_OBJETIVO)
    time.sleep(5)

    elementos = driver.find_elements(By.CSS_SELECTOR, SELECTOR_CONTENEDOR)
    print(f"Se encontraron {len(elementos)} registros.")

    for item in elementos:
        try:
            columna_a = item.find_element(By.CSS_SELECTOR, SELECTOR_DATO_A).text
            columna_b = item.find_element(By.CSS_SELECTOR, SELECTOR_DATO_B).text

            valor_limpio = "".join(filter(str.isdigit, columna_b))

            dataset_final.append({
                "tipo_obra": columna_a,
                "cantidad": valor_limpio,
                "status": 1.0
            })

        except:
            continue

    # ==========================================================
    # 4. SALIDA
    # ==========================================================
    if dataset_final:
        df = pd.DataFrame(dataset_final)
        df.to_csv("obras_coordinador.csv", index=False)

        print("Proceso exitoso ✅")
        print(df.head())
    else:
        print("No se capturaron datos")

finally:
    driver.quit()
    print("Navegador cerrado")
```
