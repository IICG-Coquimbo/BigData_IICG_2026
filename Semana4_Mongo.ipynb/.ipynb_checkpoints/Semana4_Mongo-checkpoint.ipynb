```python
from pymongo import MongoClient

client = MongoClient("mongodb://bigdata_mongodb:27017/", serverSelectionTimeoutMS=3000)
db = client["bigdata"]
coleccion = db["semana4_final"]

print("Conexión exitosa")
```

```python
documento = {
    "nombre": "Pia",
    "actividad": "Semana 4",
    "estado": "completado"
}

resultado = coleccion.insert_one(documento)

print("Insertado con ID:", resultado.inserted_id)
```

```python
for doc in coleccion.find({}, {"_id": 0}):
    print(doc)
```
