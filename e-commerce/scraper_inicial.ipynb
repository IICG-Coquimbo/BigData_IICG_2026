```python
import requests
from bs4 import BeautifulSoup

# Configuración de entrada
url = "https://books.toscrape.com/"
headers = {"User-Agent": "Mozilla/5.0"} # Para que no nos bloqueen

# 1. Petición al servidor
respuesta = requests.get(url, headers=headers)

sopa = BeautifulSoup(respuesta.text, 'html.parser')

resultados = sopa.find_all('article', class_='product_pod')
datos = [] # Lista vacia para guardar todo

for item in resultados:
    datos_item = {
        "Titulo": item.find('h3').text.strip(),
        "Precio": item.find('div', class_='product_price').text.strip()[1:7]
        }
    datos.append(datos_item)
    
print(f"Recolectado {len(datos)} datos de libros.")

for libros in datos: 
    print(libros)
```
