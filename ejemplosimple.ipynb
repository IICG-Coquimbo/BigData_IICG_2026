```python
import os
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Buscamos dónde se instaló el binario de Chromium
# Comúnmente: /usr/bin/chromium-browser o /usr/lib/chromium-browser/chromium
if os.path.exists("/usr/bin/chromium-browser"):
    options.binary_location = "/usr/bin/chromium-browser"
elif os.path.exists("/usr/bin/chromium"):
    options.binary_location = "/usr/bin/chromium"

try:
    # En muchos sistemas Linux, el driver de chromium se instala en /usr/bin/chromedriver
    service = Service("/usr/bin/chromedriver")
    driver = webdriver.Chrome(service=service, options=options)
    
    driver.get("https://www.google.com")
    print(f"¡CONECTADO EXITOSAMENTE! Título: {driver.title}")
    driver.quit()

except Exception as e:
    print(f"Error: {e}")
    print("\n--- Diagnóstico ---")
    print("¿Existe el binario?:", os.path.exists(options.binary_location))
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

options = Options()

# --- ARGUMENTOS CRÍTICOS PARA ESTABILIDAD EN LINUX ---
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage") # <--- SOLUCIONA EL ERROR DE DEVTOOLS
options.add_argument("--disable-gpu")
options.add_argument("--remote-debugging-port=9222") # Ayuda a estabilizar la conexión
options.add_argument("--window-size=1920,1080")

# La ruta que encontramos antes
options.binary_location = "/opt/conda/lib/python3.11/site-packages/seleniumbase/drivers/chrome-linux64/chrome"

try:
    # Intentamos forzar la versión del driver para que coincida con el Chrome descargado
    # Como descargaste el 146, le pedimos ese específicamente
    service = Service(ChromeDriverManager(driver_version="146.0.7680.165").install())
    
    driver = webdriver.Chrome(service=service, options=options)
    
    # Añadimos un tiempo de espera implícito por si la red es lenta
    driver.implicitly_wait(10)
    
    driver.get("https://www.google.com")
    print(f"🚀 ¡POR FIN! Conexión establecida.")
    print(f"Título: {driver.title}")
    
    driver.quit()

except Exception as e:
    print(f"❌ Error persistente: {e}")
```

```python
import os
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--disable-gpu")

# Intentamos encontrar las rutas donde Conda instaló Chromium
binary_path = "/opt/conda/bin/chromium" 
driver_path = "/opt/conda/bin/chromedriver"

# Verificamos si existen, si no, probamos rutas alternativas
if not os.path.exists(binary_path):
    binary_path = "/usr/bin/chromium-browser" # Por si ya estaba ahí

options.binary_location = binary_path

try:
    service = Service(executable_path=driver_path)
    driver = webdriver.Chrome(service=service, options=options)
    
    driver.get("https://www.google.com")
    print(f"✅ ¡CONECTADO EXITOSAMENTE! Título: {driver.title}")
    driver.quit()

except Exception as e:
    print(f"❌ Error: {e}")
    # Si falla, imprimimos dónde están los archivos para diagnosticar
    print(f"DEBUG: ¿Existe binario en {binary_path}?: {os.path.exists(binary_path)}")
```
