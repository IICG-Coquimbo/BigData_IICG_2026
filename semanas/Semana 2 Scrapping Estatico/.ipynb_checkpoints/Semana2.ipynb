```python
import requests
from bs4 import BeautifulSoup

# Configuración de entrada
url = "https://seguimientoejecucionobras.coordinador.cl/"
headers = {"User-Agent": "Mozilla/5.0"}  # Para que no nos bloqueen

# 1. Petición al servidor
respuesta = requests.get(url, headers=headers)

# 2. Convertir el HTML en un objeto de Python
sopa = BeautifulSoup(respuesta.text, 'html.parser')

# 3. Buscar los datos (usamos lo que tú encontraste)
# En tu caso: etiqueta <th> con clase "title1"
resultados = sopa.find_all('th', class_='title1')

# 4. Mostrar en pantalla
for dato in resultados:
    print(f"Encontrado: {dato.text.strip()}")
```
