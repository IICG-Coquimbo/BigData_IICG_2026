```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

options = Options()
options.binary_location = "/usr/bin/google-chrome"
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

driver.get("https://www.google.com")
print("Título:", driver.title)
driver.quit()
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

def iniciar_driver():
    options = Options()
    options.binary_location = "/usr/bin/google-chrome"
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    return webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

driver = iniciar_driver()

url = "https://finance.yahoo.com/markets/stocks/gainers/"

datos = []

try:
    print("Cargando página...")
    driver.get(url)
    time.sleep(5)

    filas = driver.find_elements(By.CSS_SELECTOR, "table tbody tr")

    print("Filas encontradas:", len(filas))

    for fila in filas:
        columnas = fila.find_elements(By.TAG_NAME, "td")

        if len(columnas) >= 6:
            datos.append({
                "simbolo": columnas[0].text,
                "nombre": columnas[1].text,
                "precio": columnas[2].text,
                "cambio": columnas[3].text,
                "porcentaje": columnas[4].text,
                "volumen": columnas[5].text
            })

    df = pd.DataFrame(datos)

    print(df.head())
    print("Total capturado:", len(df))

    df.to_csv("yahoo_gainers.csv", index=False)
    print("Archivo guardado")

except Exception as e:
    print("Error:", e)

finally:
    driver.quit()
```
