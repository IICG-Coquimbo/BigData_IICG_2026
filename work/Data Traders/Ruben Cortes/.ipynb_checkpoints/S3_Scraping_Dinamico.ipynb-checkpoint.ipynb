```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

def iniciar_driver():
    options = Options()
    options.binary_location = "/usr/bin/google-chrome"
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    return webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

driver = iniciar_driver()

url = "https://finance.yahoo.com/markets/stocks/gainers/"

datos = []

try:
    print("Cargando página...")
    driver.get(url)
    time.sleep(5)

    filas = driver.find_elements(By.CSS_SELECTOR, "table tbody tr")

    print("Filas encontradas:", len(filas))

    for fila in filas:
        columnas = fila.find_elements(By.TAG_NAME, "td")

        if len(columnas) >= 6:
            datos.append({
                "simbolo": columnas[0].text,
                "nombre": columnas[1].text,
                "precio": columnas[2].text,
                "cambio": columnas[3].text,
                "porcentaje": columnas[4].text,
                "volumen": columnas[5].text
            })

    df = pd.DataFrame(datos)

    print(df.head())
    print("Total capturado:", len(df))

    df.to_csv("yahoo_gainers.csv", index=False)
    print("Archivo guardado")

except Exception as e:
    print("Error:", e)

finally:
    driver.quit()
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

def iniciar_driver():
    options = Options()
    options.binary_location = "/usr/bin/google-chrome"
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--disable-gpu")
    return webdriver.Chrome(
        service=Service(ChromeDriverManager().install()),
        options=options
    )

def scrapear_tabla_yahoo(url, categoria):
    driver = iniciar_driver()
    datos = []

    try:
        print(f"Cargando {categoria}...")
        driver.get(url)
        time.sleep(5)

        filas = driver.find_elements(By.CSS_SELECTOR, "table tbody tr")
        print(f"Filas encontradas en {categoria}: {len(filas)}")

        for fila in filas:
            columnas = fila.find_elements(By.TAG_NAME, "td")

            if len(columnas) >= 6:
                datos.append({
                    "categoria": categoria,
                    "simbolo": columnas[0].text,
                    "nombre": columnas[1].text,
                    "precio": columnas[2].text,
                    "cambio": columnas[3].text,
                    "porcentaje": columnas[4].text,
                    "volumen": columnas[5].text
                })

    except Exception as e:
        print(f"Error en {categoria}: {e}")

    finally:
        driver.quit()

    return pd.DataFrame(datos)

# URLs
url_gainers = "https://finance.yahoo.com/markets/stocks/gainers/"
url_losers = "https://finance.yahoo.com/markets/stocks/losers/"
url_most_active = "https://finance.yahoo.com/markets/stocks/most-active/"

# Scraping
df_gainers = scrapear_tabla_yahoo(url_gainers, "gainers")
df_losers = scrapear_tabla_yahoo(url_losers, "losers")
df_most_active = scrapear_tabla_yahoo(url_most_active, "most_active")

# Guardar archivos individuales
df_gainers.to_csv("yahoo_gainers.csv", index=False)
df_losers.to_csv("yahoo_losers.csv", index=False)
df_most_active.to_csv("yahoo_most_active.csv", index=False)

# Unir todo
df_total = pd.concat([df_gainers, df_losers, df_most_active], ignore_index=True)
df_total.to_csv("yahoo_finanzas_total.csv", index=False)

print("\nResumen final:")
print("Gainers:", len(df_gainers))
print("Losers:", len(df_losers))
print("Most Active:", len(df_most_active))
print("Total general:", len(df_total))

print("\nPrimeros registros del dataset total:")
print(df_total.head())
```
